### Tratamiento del set de datos
------

Lo primero que haremos antes de realizar la exploración de datos es ordenar el set de datos y reemplazar algunos valores de filas y columnas. Puede observarse que los encabezados de las columnas están en inglés y esto puede resultar confuso para el público al que va dirigido este proyecto. Además, al ser un set de datos que clasifica a los pacientes por padecimientos o hábitos se los clasificó originalmente con 1 o 2 en cada una de las columnas que representan padecimientos, por lo que resulta conveniente cambiar esto para mejorar la comprensión del mismo a primera vista.

Entonces, desde la página de la que salió el set de datos original nos informan que en las columnas que clasifican a los pacientes por padecimientos el significado de los números 1 y 2 es 'no' y 'sí' respectivamente, por lo que sabiendo esto empezamos a realizar los cambios necesarios.

```python
# Renombro las columnas y ordeno por ID de menor a mayor.
cancer_pulmon.rename(columns={'ID':'id',
                              'first_name':'nombre',
                              'last_name':'apellido',
                              'gender':'genero',
                              'age':'edad',
                              'country':'pais',
                              'SMOKING':'fumador',
                              'YELLOW_FINGERS':'dedos_amarillos',
                              'ANXIETY':'ansiedad',
                              'PEER_PRESSURE':'presion_social',
                              'CHRONIC DISEASE':'enfermedad_cronica',
                              'FATIGUE ':'fatiga',
                              'ALLERGY ':'alergias',
                              'WHEEZING':'sibilancia',
                              'ALCOHOL CONSUMING':'consume_alcohol',
                              'COUGHING':'tos',
                              'SHORTNESS OF BREATH':'dificultad_respiratoria',
                              'SWALLOWING DIFFICULTY':'dificultad_al_tragar',
                              'CHEST PAIN':'dolor_de_pecho',
                              'LUNG_CANCER':'cancer_pulmon'}
                     ,inplace=True)
cancer_pulmon.sort_values('id',inplace=True)
cancer_pulmon.reset_index(drop=True,inplace=True)
```

```python
# Reemplazo los valores 1 y 2 de las columnas con no y sí respectivamente
columnas = ['fumador','dedos_amarillos','ansiedad','presion_social',
            'enfermedad_cronica','fatiga','alergias','sibilancia',
            'consume_alcohol','tos','dificultad_respiratoria',
            'dificultad_al_tragar','dolor_de_pecho','cancer_pulmon']
cancer_pulmon[columnas] = cancer_pulmon[columnas].replace([1,2,"NO","YES"],["no","si","no","si"])
```

### Análisis de los datos

#### Distribución por diagnóstico
-----

```python
### Agrupamos por condición de cáncer.

cancer_si_no = cancer_pulmon.groupby(by = 'cancer_pulmon').agg({'id':'count'}).reset_index()
cancer_si_no.rename(columns = {'id':'cantidad_pacientes'}, inplace = True) ### Renombro la columna
cancer_si_no['porcentaje'] = np.round((cancer_si_no['cantidad_pacientes']/2000)*100,0) ### Genero una columna que me muestre la cantidad en porcentaje
cancer_si_no
```

Con la agrupación se generó una pequeña lista que nos permite conocer la cantidad de pacientes de este set de datos que tienen cáncer y la cantidad que no tienen cáncer. Se puede observar que la mayoría de los pacientes presentan un diagnóstico positivo, lo que representa un 87% del total del set de datos. Esto podemos visualizarlo mejor en un gráfico de torta.

```python
fig, ax = plt.subplots(figsize=(10, 6))

sizes = [87, 13]

labels = ['Positivo', 'Negativo']

ax.pie(sizes, labels=labels, autopct='%1.1f%%',
       colors = ['dodgerblue', 'lightblue'], textprops = {'fontsize':14})

plt.title("Distribución de los pacientes por diagnóstico", fontsize = 20)
```

Con el gráfico resulta mucho más visible el porcentaje de pacientes con cáncer que forman parte de nuestro set de datos. Esto es una señal clara de lo sesgada que está la muestra, siendo que la gran mayoría posee un diagnóstico positivo.

#### Relación entre género y diagnóstico
-----

Ahora podría interesarnos observar cómo se distribuyen los mismos en cuanto a género y diagnóstico, para adentrarnos un poco más a entender si existen patrones por género también. Primero observemos en una tabla los diferentes géneros que tenemos, de la siguiente manera:

```python
### Agrupo por género para observar los géneros y la cantidad de pacientes

generos = cancer_pulmon.groupby(by = 'genero').agg({'id':'count'}).reset_index()
generos.rename(columns = {'id':'cant_pacientes'}, inplace = True) ### Renombro la columna
generos['porcentaje'] = np.round((generos['cant_pacientes']/2000)*100, 1) ### Porcentaje del total
generos.sort_values(by = 'cant_pacientes', ascending = False, inplace = True) ### Ordeno de mayor a menor
generos.reset_index(drop = True, inplace = True)
generos
```

Podemos observar que tenemos en total 8 géneros en nuestro set de datos, siendo casi el 90% la suma entre hombres y mujeres. Lo que podemos hacer a continuación es agrupar a los demás géneros para facilitar los análisis, como se muestra a continuación:

```python
### Reemplazamos los géneros que representan poco volumen

cancer_pulmon.replace(['Non-binary', 'Polygender', 'Agender', 'Genderfluid',
                        'Bigender', 'Genderqueer'], 'Other', inplace = True)

generos1 = cancer_pulmon.groupby(by = 'genero').agg({'id':'count'}).reset_index()
generos1.rename(columns = {'id':'cant_pacientes'}, inplace = True) ### Renombro la columna
generos1['porcentaje'] = np.round((generos1['cant_pacientes']/2000)*100, 1) ### Porcentaje del total
generos1.sort_values(by = 'cant_pacientes', ascending = False)### Ordeno de mayor a menor
```

Ahora podemos observar con mayor claridad los diferentes géneros, quedando así 3 grandes grupos: hombres, mujeres y otros, siendo este último una muestra mucho más importante para los análisis.

Lo siguiente que podemos observar es cómo se distribuyen estas cantidades entre los dos diagnósticos de los que disponemos.

```python
genero_diag = cancer_pulmon.groupby(by = ['genero', 'cancer_pulmon']).agg({'id':'count'}).reset_index()
genero_diag.rename(columns = {'id':'cant_pacientes', 'cancer_pulmon':'Cáncer de pulmón'}, inplace = True)
genero_diag.sort_values(by = 'cant_pacientes', ascending = False)

### Armamos un pivot con los datos para que resulte mejor visualmente
genero_diag_pivot = genero_diag.pivot(index = 'genero', columns = 'Cáncer de pulmón', values = 'cant_pacientes')
genero_diag_pivot.sort_values(['si', 'no'], ascending = False, inplace = True)
genero_diag_pivot.reset_index(inplace = True)

### Calculo también los porcentajes 
genero_diag_pivot['pct_negativo'] = np.round((genero_diag_pivot['no']/2000)*100, 1)
genero_diag_pivot['pct_positivo'] = np.round((genero_diag_pivot['si']/2000)*100, 1)
genero_diag_pivot.rename(columns = {'no':'negativo', 'si':'positivo'})
```

Con esta tabla podemos ver con mayor claridad la cantidad de pacientes por género y diagnóstico, siendo la mayoría con diagnóstico positivo, como ya habíamos confirmado. Lo que nos muestran además los datos es que por poco los que más propensos parecieran ser al cáncer son las mujeres, más que los hombres, siendo un 40% del total de pacientes.

Veamos estos datos graficados, para que sea más visual la interpretación de los mismos.

```python
### Reordenamos el dataframe para poder graficarlo
genero_diag.sort_values(by = 'cant_pacientes', ascending = False, inplace = True)
genero_diag.reset_index(drop = True, inplace = True)
### También podemos reemplazar los nombres de los géneros en inglés por los que son en español
genero_diag.replace(['Female', 'Male','Other'], ['Femenino', 'Masculino', 'Otro'], inplace = True)
```

```python
plt.figure(figsize = (10,6))

# paleta = {'lightblue', 'dodgerblue'}

bar1 = sns.barplot(x = 'genero', y = 'cant_pacientes', hue = 'Cáncer de pulmón',
                   data = genero_diag, palette = {'si':'dodgerblue', 'no':'lightblue'}, edgecolor = 'dimgray')

### Acá aplico los valores del pivot para agregar el valor a cada barra
plt.bar_label(bar1.containers[0], labels = genero_diag_pivot['si'], label_type = 'edge',
             fontsize=12)
plt.bar_label(bar1.containers[1], labels = genero_diag_pivot['no'], label_type = 'edge',
             fontsize=12)

plt.title("Cantidad de pacientes por género y diagnóstico", fontsize=16)
plt.xlabel(None)
plt.ylabel("Cantidad de pacientes", fontsize=10)
plt.tick_params(axis='both', labelsize=10)
plt.ylim(0, 900)
```

El gráfico nos permite observar mejor los datos y la información que nos proporcionan. Confirmamos con esto que los pacientes más propensos son mujeres pero los hombres también forman un volumen importante en los análisis, representando el 45% y el 44% respectivamente. Esto refuta la hipótesis de que los hombres son el género más propenso a padecer cáncer.

#### Densidad de distribución por edad y diagnóstico
------

Lo siguiente que nos interesa saber es cómo se distribuyen los pacientes por edad, género y diagnóstico. Lo que podemos hacer para conocer esto es una distribución de densidad. A continuación se muestra cómo se logra esto:

```python
### En esta parte procedemos a separar los diferentes datos de los pacientes
### por género y diagnóstico

### Female
female = cancer_pulmon[cancer_pulmon['genero'] == 'Female']
female.reset_index(drop = True, inplace = True)
# Sin cáncer
female_no = female[female['cancer_pulmon'] == 'no']
female_no.reset_index(drop = True, inplace = True)
# Con cáncer
female_si = female[female['cancer_pulmon'] == 'si']
female_si.reset_index(drop = True, inplace = True)

### Male
male = cancer_pulmon[cancer_pulmon['genero'] == 'Male']
male.reset_index(drop = True, inplace = True)
# Sin cáncer
male_no = male[male['cancer_pulmon'] == 'no']
male_no.reset_index(drop = True, inplace = True)
# Con cáncer
male_si = male[male['cancer_pulmon'] == 'si']
male_si.reset_index(drop = True, inplace = True)

### Other
other = cancer_pulmon[cancer_pulmon['genero'] == 'Other']
other.reset_index(drop = True, inplace = True)
# Sin cáncer
other_no = other[other['cancer_pulmon'] == 'no']
other_no.reset_index(drop = True, inplace = True)
# Con cáncer
other_si = other[other['cancer_pulmon'] == 'si']
other_si.reset_index(drop = True, inplace = True)
```

```python
### Grafico los datasets por género y diagnóstico

fig, ax = plt.subplots(ncols = 3, figsize = (15, 5),
                       sharex = True, sharey = True)

fig.suptitle("Densidad por género y diagnóstico", fontsize = 16)

paleta = ['lightblue', 'dodgerblue']

### Female ###
sns.kdeplot(ax = ax[0], data = female_no['edad'], fill = True,
            color = paleta[0])
sns.kdeplot(ax = ax[0], data = female_si['edad'], fill = True,
            color = paleta[1])

ax[0].set_title("Femenino", fontsize = 12,
             loc = 'center')
ax[0].set_ylabel("Densidad")
ax[0].set_xlabel(None)

### Male ###
sns.kdeplot(ax = ax[1], data = male_no['edad'], fill = True,
            color = paleta[0])
sns.kdeplot(ax = ax[1], data = male_si['edad'], fill = True,
            color = paleta[1])

ax[1].set_title("Masculino", fontsize = 12,
             loc = 'center')
ax[1].set_ylabel(None)
ax[1].set_xlabel("Edad")

### Other ###
sns.kdeplot(ax = ax[2], data = other_no['edad'], fill = True,
            color = paleta[0], label = 'Negativo')
sns.kdeplot(ax = ax[2], data = other_si['edad'], fill = True,
            color = paleta[1], label = 'Positivo')
ax[2].set_title("Otro", fontsize = 12, loc = 'center')
ax[2].set_xlabel(None)
ax[2].legend()
```

Viendo a los pacientes de esta manera se tiene un mejor entendimiento de las distribuciones. Observando los gráficos de densidad se puede resaltar que no hay diferencias significativas de los pacientes **mujeres** y **hombres** que con distinto diagnóstico, ya que los gráficos celeste y azul casi se solapan en cada género, por lo que no hay distinción en cuanto a la edad en que un paciente de estos géneros puede contraer cáncer. Por otro lado en el gráfico de 'Otro' se puede observar que existe un pico para los pacientes de ese grupo con diagnóstico negativo y los positivos se distribuyen de manera más uniforme, presentando una leve pendiente a medida que crece la edad.

Lo que nos dice esto es que para los géneros **femenino** y **masculino** no hay un rango de edad tan definido como se planteó en las hipótesis sino que resulta ser bastante homogénea la edad en que los pacientes presentaron cáncer.
```python
### Importamos esto para transformar las variables de texto en número
from sklearn import preprocessing
```

```python
### Armo el modelo de Encoding para las variables de texto
encoding = preprocessing.LabelEncoder()
```

```python
### Armo un nuevo set de datos en base al original
cancer_pulmon_heatmap = cancer_pulmon.drop(columns = ['id', 'nombre', 'apellido', 
                                                      'genero', 'edad', 'pais'])
cancer_pulmon_heatmap.head() ### Elimino las columnas que no necesito
```

#### Prueba de modelos y métricas
------

Probamos a continuación 3 diferentes modelos de Clasificación de Machine Learning, que son: Árbol de decisión, KNN y Regresión Logística. Una breve descripción de cada uno:

1. Árbol de decisión: permite evaluar mediante una representación gráfica los posibles resultados, costos y consecuencias de una decisión compleja.
2. KNN: utiliza la proximidad para hacer clasificaciones o predicciones sobre la agrupación de un punto de datos individual.
3. Regresión logística: técnica estadística para predecir variables categóricas mediante variables predictoras.

Para entrenar y testear estos modelos tomamos las variables seleccionadas de la matriz de correlación. Luego para evaluar la performance de los modelos seleccionados se utilizarán las métricas F1 Score y la Matriz de Confusión, y se aplicarán tanto a la variable de entrenamiento como a la de testeo. Una breve descripción de cada métrica:

1. Matriz de Confusión: permite evaluar el desempeño de un algoritmo de aprendizaje supervisado. Permite observar los aciertos y errores del modelo.
2. F1 Score: medida general del desempeño de un modelo. Si lo que devuelve es un valor cercano a 1 quiere decir que tenemos pocos falsos positivos y pocos falsos negativos.
3. Accuracy: porcentaje total de valores correctamente clasificados, tanto positivos como negativos.

Luego, al aplicarlas tanto en entrenamiento como en testeo podemos validar si nuestros modelos predicen con overfitting o underfitting, lo cual no es bueno para determinar si la performance de un modelo es buena.

Entonces, para comenzar a realizar esto se importarán las librerías necesarias: trabajaremos esta sección con sklearn, de la cual podemos importar tanto modelos como métricas.

```python
### Importo librerías necesarias para entrenar y para probar los modelos

from sklearn.model_selection import train_test_split # Función para entrenar y testear

### Modelos
# Árbol de decisión
from sklearn.tree import DecisionTreeClassifier
# Regresión Logística
from sklearn.linear_model import LogisticRegression
# Vecinos cercanos o KNN
from sklearn.neighbors import KNeighborsClassifier

#### Métricas
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score
from sklearn import metrics
from sklearn.model_selection import cross_val_score

### Importo la librería warnings para no tener carteles molestos
import warnings as wn
```

Armo los dos pares de variables:

1. De entrenamiento (train): estas servirán para entrenar los modelos de ML.
2. De testeo (test): estas servirán para probar si al calcular las métricas obtengo overfitting o underfitting.

```python
### Genero los datos de entrenamiento para los modelos
Y = cancer_enc[y_col]

X = cancer_enc[x_cols]

### Variables de entrenamiento

x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size = 0.2, ## Porcentaje de los datos que será para testear los modelos
                                                    random_state = 50)
```

```python
##### Definición de los modelos

# Arbol de decisión
arbol = DecisionTreeClassifier(max_depth = 3)

# Regresión Logística
regresion = LogisticRegression(penalty = 'l2',
                               solver = 'liblinear',
                               multi_class = 'ovr')

# KNN
knn = KNeighborsClassifier()
```

```python
### Armo una lista con los modelos

modelos = [arbol, regresion, knn]
```

```python
### Creo un for que pruebe en bucle cada modelo y las métricas

for modelo in modelos:

    wn.filterwarnings('ignore')

    print(f'El modelo que usaremos a continuación es {modelo} ->')

    # Entreno los modelos
    modelo.fit(x_train, y_train)

    # Definimos el valor de prediccion
    y_train_pred = modelo.predict(x_train)
    y_pred = modelo.predict(x_test)
    
    ### Calculo las métricas con Accuracy
    accuracy_train = accuracy_score(y_train, y_train_pred)
    accuracy_test = accuracy_score(y_test, y_pred)

    # Calculo las métricas con F1 Score
    f1_train = np.mean(cross_val_score(modelo, x_train, y_train, cv = 10, scoring = 'f1'))
    f1_test = np.mean(cross_val_score(modelo, x_test, y_test, cv = 10, scoring = 'f1'))

    # Imprimimos los valores
    print(f'La métrica Accuracy del modelo de entrenamiento es de: {accuracy_train}')
    print(f'La métrica Accuracy del modelo de testeo es de: {accuracy_test}')
    print(f'La métrica F1 Score del modelo de entrenamiento es de: {f1_train}')
    print(f'La métrica F1 Score del modelo de testeo es de: {f1_test}')

    # Uso además la matriz de confusión
    cm = confusion_matrix(y_test, y_pred)

    # Graficamos la matriz de confusión
    sns.heatmap(cm,
                annot = True,
                fmt = 'g',
                xticklabels = ['Sí', 'No'],
                yticklabels = ['Sí', 'No'],
               cmap = 'Blues')
    plt.ylabel('Predicción', fontsize = 13)
    plt.xlabel('Real', fontsize = 13)
    plt.title('Matriz de confusión', fontsize = 17)
    plt.show()
```

#### Mejora de modelos con PCA
------

Utilizando los modelos entrenados con el nuevo set de datos y sabiendo lo que sabemos de las métricas podemos preguntarnos si es posible mejorar la performance de los modelos aplicando PCA. Se trata de lo siguiente: PCA es un método estadístico cuya utilidad radica en la reducción de la dimensionalidad de la base de datos, es decir minimizar la cantidad de variables adecuadas para predecir los datos.

Ahora importamos las librerías necesarias para trabajar esto:



```python
# Libreria para estandarizar los datos
from sklearn.preprocessing import StandardScaler

# Importamos el modelo de PCA
from sklearn.decomposition import PCA
```

```python
# Normalizamos los datos
sc = StandardScaler()
X_train = sc.fit_transform(x_train) ### Estandarizamos
X_test = sc.fit_transform(x_test) ### Estandarizamos

# Establecemos el pca
pca = PCA()

# Aplicamos el pca sobre nuestras variables
X_train = pca.fit_transform(X_train)
X_test = pca.fit_transform(X_test)

# Analizamos las variables
explainend_variance = pca.explained_variance_ratio_
explainend_variance
```

El output que se obtuvo como un array indica el porcentaje de peso que tiene cada variable respecto a la varianza que tenemos. Lo que sigue a continuación es evaluar los modelos con PCA.

```python
# Crearemos un bucle for que evalue cada uno de los modelos por separados para saber cual de todos ellos es el mejor para nuestro modelo.

for modelo in modelos:
    
    print(f'#######################################################################')
    print(f'El modelo que usaremos a continuacion es {modelo} ->')
    print(f'#######################################################################')
    
    # Definimos el numero de componentes a implementar
    cant_componentes = 10
    
    # Emplearemos un modelo en base a esta informacion
    pca = PCA(n_components = cant_componentes)
    X_train = pca.fit_transform(X_train)
    X_test = pca.transform(X_test)
    
    # Definimos con que entrenar los modelos
    modelo.fit(X_train, y_train)
    
    # Definimos el valor de prediccion
    y_train_pred = modelo.predict(X_train)
    y_pred = modelo.predict(X_test)
    
    ### Calculo las métricas con Accuracy
    accuracy_train = accuracy_score(y_train, y_train_pred)
    accuracy_test = accuracy_score(y_test, y_pred)
    
    # Calculo las métricas con F1 Score
    f1_train = np.mean(cross_val_score(modelo, X_train, y_train, cv=10, scoring='f1'))
    f1_test = np.mean(cross_val_score(modelo, X_test, y_test, cv=10, scoring='f1'))

    # Imprimimos los valores
    print(f'La métrica Accuracy del modelo de entrenamiento es de: {accuracy_train}')
    print(f'La métrica Accuracy del modelo de testeo es de: {accuracy_test}')
    print(f'La métrica F1 Score del modelo de entrenamiento es de: {f1_train}')
    print(f'La métrica F1 Score del modelo de testeo es de: {f1_test}')
    
    # Creamos la matriz de confusión
    cm = confusion_matrix(y_test, y_pred)

    # Graficamos la matriz de confusión
    sns.heatmap(cm,
                annot=True,
                fmt='g',
                xticklabels=['Sí', 'No'],
                yticklabels=['Sí', 'No'],
               cmap = 'Blues')
    plt.ylabel('Predicción',fontsize=13)
    plt.xlabel('Real',fontsize=13)
    plt.title('Matriz de confusión',fontsize=17)
    plt.show()
```

Aunque descartemos PCA aún existe la necesidad de validar los modelos que se entrenaron por lo que utilizaremos una técnica llamada Cross Validation, que nos dice lo siguiente: es un método que consiste en evaluar y probar el rendimiento de un modelo de machine learning, con el fin de encontrar un mejor modelo rápidamente. Entonces aplicaremos esto a los modelos de ML con el set de datos de 50/50 y utilizando las métricas F1 Score y Accuracy.
