#### Acomodar el set de datos y "ponerlo lindo"

```python
# Renombro las columnas y ordeno por ID de menor a mayor.
df_cancer_pulmon.rename(columns={'ID':'id',
                              'first_name':'nombre',
                              'last_name':'apellido',
                              'gender':'genero',
                              'age':'edad',
                              'country':'pais',
                              'SMOKING':'fumador',
                              'YELLOW_FINGERS':'dedos_amarillos',
                              'ANXIETY':'ansiedad',
                              'PEER_PRESSURE':'presion_social',
                              'CHRONIC DISEASE':'enfermedad_cronica',
                              'FATIGUE ':'fatiga',
                              'ALLERGY ':'alergias',
                              'WHEEZING':'sibilancia',
                              'ALCOHOL CONSUMING':'consume_alcohol',
                              'COUGHING':'tos',
                              'SHORTNESS OF BREATH':'dificultad_respiratoria',
                              'SWALLOWING DIFFICULTY':'dificultad_al_tragar',
                              'CHEST PAIN':'dolor_de_pecho',
                              'LUNG_CANCER':'cancer_pulmon'}
                     ,inplace=True)
df_cancer_pulmon.sort_values('id',inplace=True)
df_cancer_pulmon.reset_index(drop=True,inplace=True)
```

Ahora armo una agrupación del dataset por género y por cáncer de pulmón.

```python
### Cantidad de pacientes por género y con cáncer o no
df_genero = df_cancer_pulmon.groupby(by = ['genero', 'cancer_pulmon']).agg({'id':'count'}).reset_index()
df_genero.rename(columns={'id':'cant_pacientes'},inplace=True)
df_genero.sort_values('cant_pacientes',ascending=False,inplace=True)
df_genero.reset_index(drop=True,inplace=True)
df_genero.rename(columns = {'cancer_pulmon':'Cáncer de pulmón'}, inplace = True)

### Para que la tabla que salga de esta agrupación se vea bien
### aplico un pivot a la tabla.

df_agrup = df_genero.pivot(index = 'genero', columns = 'Cáncer de pulmón', values = 'cant_pacientes')
df_agrup.sort_values(['si', 'no'], ascending = False, inplace = True)
df_agrup.reset_index(inplace = True)
df_agrup
```

Se observa en la tabla que **la mayoría de los pacientes están diagnosticados con cáncer**, esto quiere decir que la muestra de pacientes está desbalanceada, siendo más casos positivos que negativos en el set de datos.

#### Gráfico de barras comparativo: género y condición sí o no de cáncer

```python
### Grafico los datos obtenidos del group by

plt.figure(figsize = (10,6))

paleta = {'springgreen', 'mediumpurple'}

bar1 = sns.barplot(x = 'genero', y = 'cant_pacientes', hue = 'Cáncer de pulmón',
                   data = df_genero, palette = paleta, edgecolor = 'dimgray')


### Acá aplico los valores del pivot para agregar el valor a cada barra
plt.bar_label(bar1.containers[0], labels = df_agrup['si'], label_type = 'edge',
             fontsize=12)
plt.bar_label(bar1.containers[1], labels = df_agrup['no'], label_type = 'edge',
             fontsize=12)

plt.title("Cantidad de pacientes por género y cáncer", fontsize=16)
plt.xlabel(None)
plt.ylabel("Cantidad de pacientes", fontsize=10)
plt.tick_params(axis='both', labelsize=10)
plt.ylim(0, 900)
```

#### Gráfico de densidad: cantidad de pacientes por edad y por género, aplicando condición de cáncer sí o no

Ahora que ya se analizó el dataset para conocer la distribución por género y condición de cáncer, muestro a través de gráficos de densidad la distribución de los clientes agrupados por edad y segmentado por la condición de cáncer. Entonces en el gráfico por cada género se van a observar dos densidades: una que corresponde a la distribución de los pacientes que no tienen cáncer y otra que corresponde a los pacientes que sí tienen cáncer. Para esto armo diferentes datasets que separen los pacientes por género y luego que los separe por la condición cáncer sí/no.

```python
### Separo los datos por género

### Female
female = df_cancer_pulmon[df_cancer_pulmon['genero'] == 'Female']
female.reset_index(drop = True, inplace = True)
# Sin cáncer
female_no = female[female['cancer_pulmon'] == 'no']
female_no.reset_index(drop = True, inplace = True)
# Con cáncer
female_si = female[female['cancer_pulmon'] == 'si']
female_si.reset_index(drop = True, inplace = True)

### Male
male = df_cancer_pulmon[df_cancer_pulmon['genero'] == 'Male']
male.reset_index(drop = True, inplace = True)
# Sin cáncer
male_no = male[male['cancer_pulmon'] == 'no']
male_no.reset_index(drop = True, inplace = True)
# Con cáncer
male_si = male[male['cancer_pulmon'] == 'si']
male_si.reset_index(drop = True, inplace = True)

### Other
other = df_cancer_pulmon[df_cancer_pulmon['genero'] == 'Other']
other.reset_index(drop = True, inplace = True)
# Sin cáncer
other_no = other[other['cancer_pulmon'] == 'no']
other_no.reset_index(drop = True, inplace = True)
# Con cáncer
other_si = other[other['cancer_pulmon'] == 'si']
other_si.reset_index(drop = True, inplace = True)
```

```python
### Grafico los datasets por género y condición

fig, ax = plt.subplots(ncols = 3, figsize = (10, 5),
                       sharex = True, sharey = True)

fig.suptitle("Densidad por género y condición cáncer", fontsize = 16)

paleta = ['mediumpurple', 'springgreen']

### Female ###
sns.kdeplot(ax = ax[0], data = female_no['edad'], fill = True,
            color = paleta[0])
sns.kdeplot(ax = ax[0], data = female_si['edad'], fill = True,
            color = paleta[1])

ax[0].set_title("Female", fontsize = 12,
             loc = 'center')
ax[0].set_ylabel("Densidad")
ax[0].set_xlabel(None)

### Male ###
sns.kdeplot(ax = ax[1], data = male_no['edad'], fill = True,
            color = paleta[0])
sns.kdeplot(ax = ax[1], data = male_si['edad'], fill = True,
            color = paleta[1])

ax[1].set_title("Male", fontsize = 12,
             loc = 'center')
ax[1].set_ylabel(None)
ax[1].set_xlabel("Edad")

### Other ###
sns.kdeplot(ax = ax[2], data = other_no['edad'], fill = True,
            color = paleta[0], label = 'Sin cáncer')
sns.kdeplot(ax = ax[2], data = other_si['edad'], fill = True,
            color = paleta[1], label = 'Con cáncer')
ax[2].set_title("Other", fontsize = 12, loc = 'center')
ax[2].set_xlabel(None)
ax[2].legend()

```

Observando los gráficos de densidad se puede determinar que no hay diferencias significativas de los pacientes 'female' y 'male' que presentan o no cáncer, ya que los gráficos casi se solapan, por lo que no hay distinción en cuanto a la edad en que un paciente de estos géneros puede contraer cáncer. Mientras que en el gráfico de 'other' se puede observar que existe un pico para los pacientes de ese grupo que no tienen cáncer y los que sí tienen cáncer se distribuyen un poco más de manera uniforme, presentando una leve pendiente a medida que crece la edad.

#### Gráficos de mapa: cantidad de pacientes con cáncer por país vs cantidad de pacientes sin cáncer por país

Ahora que se desglosó la edad de los pacientes quiero conocer la nacionalidad de los mismos, observando las distribuciones globales entre pacientes con y sin cáncer.

Para poder trabajar con mapas puedo utilizar la librería geopandas, que trabaja con polígonos, por lo que la instalo a continuación.

```python
### Instalo la librería geopandas
!pip install geopandas
```

```python
### Importo la librería
import geopandas as gpd
```

```python
### Nombro una variable que contenga la url donde se encuentra el archivo que necesito
file_countries = "C:/Users/175544/Desktop/Curso Coderhouse/Data Science/Avances PF/countries.geo.json"
```

```python
### Cargo el archivo como un geodataframe
geo_countries = gpd.read_file(file_countries)
geo_countries.rename(columns={'name':'pais','id':'abrev'},inplace=True)
geo_countries.head()
```

```python
### Cruzo la base que tiene los polígonos con la base de pacientes a través de la columna 'pais'
cruce_dataset = pd.merge(geo_countries, df_cancer_pulmon, how='left', on='pais')
cruce_dataset.head()
```

```python
### Separo a los pacientes con cáncer de los que no tiene cáncer para observarlos de forma comparativa en los mapas
pacientes_no = cruce_dataset[cruce_dataset['cancer_pulmon']=='no']
pacientes_no.reset_index(drop = True, inplace = True)

pacientes_si = cruce_dataset[cruce_dataset['cancer_pulmon']=='si']
pacientes_si.reset_index(drop = True, inplace = True)

### Armo un group by del país y por polígono y cuento la cantidad de pacientes por país

paises_no = pacientes_no.groupby(by = ['pais', 'abrev', 'geometry']).agg({'nombre':'count'}).reset_index()
paises_no.rename(columns={'nombre':'cant_pacientes'}, inplace = True)

paises_si = pacientes_si.groupby(by = ['pais', 'abrev', 'geometry']).agg({'nombre':'count'}).reset_index()
paises_si.rename(columns={'nombre':'cant_pacientes'}, inplace = True)

### Luego del group by vuelvo a transformar el dataframe a un geodataframe
geo_paises_no = gpd.GeoDataFrame(paises_no)
geo_paises_si = gpd.GeoDataFrame(paises_si)
```

### Prueba de modelos y Métricas

En esta sección pruebo los modelos de Clasificación de Machine Learning que se practicaron en clase: Árbol de decisión, KNN y Regresión Logística. Para esto tomo las variables que se listaron arriba y entreno a los modelos. Luego se probarán las métricas F1 Score y la Matriz de Confusión para observar si las predicciones son correctas o no.

Entonces para poder realizar estos pasos se requiere importar algunas librerías de scikilearn, que se muestran a continuación.

```python
### Importo librerías necesarias para entrenar y para probar los modelos
# Para entrenar y testear
from sklearn.model_selection import train_test_split
# Árbol de decisión
from sklearn.tree import DecisionTreeClassifier
# Regresión Logística
from sklearn.linear_model import LogisticRegression
# Vecinos cercanos o KNN
from sklearn.neighbors import KNeighborsClassifier

#### Métricas
from sklearn.metrics import f1_score, confusion_matrix
from sklearn import metrics
from sklearn.model_selection import cross_val_score

### Importo la librería warnings para no tener carteles molestos
import warnings as wn
```

Armo dos pares de variables:

1. De entrenamiento (train): estas servirán para entrenar los modelos de ML.
2. De testeo (test): estas servirán para probar si al calcular las métricas obtengo overfitting o underfitting.

```python
### Genero los datos de entrenamiento para los modelos
Y = df_cancer_enc1[y_col]

X = df_cancer_enc1[x_cols]

### Variables de entrenamiento

x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size = 0.2, ## Porcentaje de los datos que será para testear los modelos
                                                    random_state = 50)
```

```python
##### Definición de los modelos
# Arbol de decisión
arbol = DecisionTreeClassifier(max_depth = 3)

# Regresión Logística
regresion = LogisticRegression(penalty = 'l2',
                               solver = 'liblinear',
                               multi_class = 'ovr')

# KNN
knn = KNeighborsClassifier()
```

```python
### Armo una lista con los modelos

modelos = [arbol, regresion, knn]
```

```python
### Creo un for que pruebe en bucle cada modelo y las métricas

for modelo in modelos:

    wn.filterwarnings('ignore')

    print(f'El modelo que usaremos a continuacion es {modelo} ->')

    # Entreno los modelos
    modelo.fit(x_train, y_train)

    # Definimos el valor de prediccion
    y_train_pred = modelo.predict(x_train)
    y_pred = modelo.predict(x_test)

    # Calculo las métricas con F1 Score
    f1_train = np.mean(cross_val_score(modelo, x_train, y_train, cv=10, scoring='f1'))
    f1_test = np.mean(cross_val_score(modelo, x_test, y_test, cv=10, scoring='f1'))

    # Imprimimos los valores
    print(f'La métrica F1 Score del modelo de entrenamiento es de: {f1_train}')
    print(f'La métrica F1 Score del modelo de testeo es de: {f1_test}')

    # Uso además la matriz de confusión
    cm = confusion_matrix(y_test, y_pred)

    # Graficamos la matriz de confusión
    sns.heatmap(cm,
                annot=True,
                fmt='g',
                xticklabels=['No', 'Sí'],
                yticklabels=['No', 'Sí'])
    plt.ylabel('Predicción',fontsize=13)
    plt.xlabel('Real',fontsize=13)
    plt.title('Matriz de confusión',fontsize=17)
    plt.show()
```
