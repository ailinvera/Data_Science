Se observa que el set de datos es aparentemente una lista de pacientes con nombre completo, género, edad, país de origen y diferentes hábitos o padecimientos, y adicionalmente una columna que parece decirnos si el paciente en cuestión padece o no cáncer, por lo que la descripción original de los datos se puede considerar bastante acertada.

Previo a lo dicho en el párrafo anterior lo que tenemos que entender es si nuestro set de datos es adecuado para realizar cualquier tipo de análisis que genere información, lo que nos lleva a tener que realizar una limpieza de los datos (en inglés Data Wrangling). Esto implica realizar un pequeño análisis de los datos pero a nivel de calidad de los datos: tipo de datos adecuado, valores duplicados, valores nulos o atípicos.

Lo primero que haremos antes de realizar la exploración de datos es ordenar el set de datos y reemplazar algunos valores de filas y columnas. Puede observarse que los encabezados de las columnas están en inglés y esto puede resultar confuso para el público al que va dirigido este proyecto. Además al ser un set de datos que clasifica a los pacientes por padecimientos o hábitos se los clasificó originalmente con 1 o 2 en cada una de las columnas que representan padecimientos, por lo que resulta conveniente cambiar esto para mejorar la comprensión del mismo a primera vista.

Entonces, desde la página de la que salió el set de datos original nos informan que en las columnas que clasifican a los pacientes por padecimientos el significado de los números 1 y 2 es 'no' y 'sí' respectivamente, por lo que sabiendo esto empezamos a realizar los cambios necesarios.

```python
# Renombro las columnas y ordeno por ID de menor a mayor.
cancer_pulmon.rename(columns={'ID':'id',
                              'first_name':'nombre',
                              'last_name':'apellido',
                              'gender':'genero',
                              'age':'edad',
                              'country':'pais',
                              'SMOKING':'fumador',
                              'YELLOW_FINGERS':'dedos_amarillos',
                              'ANXIETY':'ansiedad',
                              'PEER_PRESSURE':'presion_social',
                              'CHRONIC DISEASE':'enfermedad_cronica',
                              'FATIGUE ':'fatiga',
                              'ALLERGY ':'alergias',
                              'WHEEZING':'sibilancia',
                              'ALCOHOL CONSUMING':'consume_alcohol',
                              'COUGHING':'tos',
                              'SHORTNESS OF BREATH':'dificultad_respiratoria',
                              'SWALLOWING DIFFICULTY':'dificultad_al_tragar',
                              'CHEST PAIN':'dolor_de_pecho',
                              'LUNG_CANCER':'cancer_pulmon'}
                     ,inplace=True)
cancer_pulmon.sort_values('id',inplace=True)
cancer_pulmon.reset_index(drop=True,inplace=True)
```

```python
# Reemplazo los valores 1 y 2 de las columnas con no y sí respectivamente
columnas = ['fumador','dedos_amarillos','ansiedad','presion_social',
            'enfermedad_cronica','fatiga','alergias','sibilancia',
            'consume_alcohol','tos','dificultad_respiratoria',
            'dificultad_al_tragar','dolor_de_pecho','cancer_pulmon']
cancer_pulmon[columnas] = cancer_pulmon[columnas].replace([1,2,"NO","YES"],["no","si","no","si"])
```

```python
cancer_pulmon.head()
```

Ahora el set de datos tiene más sentido para cualquier lector, por lo que a continuación nos resta averiguar si el set de datos requiere una limpieza.

Veremos si el set de datos tiene valores nulos en sus columnas.

```python
### Podemos saber si existen nulos con la función siguiente
cancer_pulmon.info()
```

Esta función nos muestra no sólo la cuenta por columna de los valores no nulos sino también el tipo de dato de cada columna. Puede observarse que el set de datos no tiene ningún valor nulo, ya que cuenta por cada columna 2000 valores no nulos, que es el total de filas del set de datos.

Ahora observaremos si existen valores duplicados, considerando lo siguiente: sabemos que el set de datos es una lista de pacientes con diferentes características como la edad, género y nacionalidad, además de las ya citadas enfermedades o hábitos. Lo que podemos preguntarnos con esto es si entre los pacientes existen valores duplicados, es decir registros de pacientes que están listados dos veces. Lo que podemos hacer es eliminar duplicados y chequear la cantidad de pacientes de la siguiente manera:

```python
### Eliminamos duplicados a priori y contamos la cantidad que queda
cancer_pulmon.drop_duplicates(subset = ['nombre', 'apellido'])['id'].count()
```

Lo que nos dice esto es que no existen valores duplicados de pacientes por nombre y apellido, lo que podemos interpretar como pacientes únicos.

Lo que nos resta entender es si existen valores atípicos. Los valores atípicos son aquellos que no se ajustan a la norma como los demás valores. Para considerar esto pensemos un momento en nuestro set de datos: como ya se dijo, el set de datos consta de una lista de pacientes con diferentes características y que además están clasificados por padecimientos. ¿Cómo podría alguno de ellos representar un valor atípico? La respuesta es que ninguno podría ser un valor atípico ya que no existe o no tenemos pensada una norma para los mismos.

Con estos análisis breves de la calidad del set de datos concluye la parte de limpieza y comenzamos con la parte de análisis de los datos para extraerles información. Como se calculó más arriba sabemos que disponemos de 2000 pacientes en nuestro set de datos. Ahora lo que podríamos observar es la cantidad de los mismos que tiene cáncer o no, como se calcula a continuación.

```python
### Agrupamos por condición de cáncer.

cancer_si_no = cancer_pulmon.groupby(by = 'cancer_pulmon').agg({'id':'count'}).reset_index()
cancer_si_no.rename(columns = {'id':'cantidad_pacientes'}, inplace = True) ### Renombro la columna
cancer_si_no['porcentaje'] = np.round((cancer_si_no['cantidad_pacientes']/2000)*100,0) ### Genero una columna que me muestre la cantidad en porcentaje
cancer_si_no
```

Con la agrupación se generó una pequeña lista que nos permite conocer la cantidad de pacientes de este set de datos que tienen cáncer y la cantidad que no tiene cáncer. Se puede observar que la mayoría de los pacientes presentan cáncer, siendo un 87% del total del set de datos. Esto podemos visualizarlo mejor en un gráfico de torta.

```python
fig, ax = plt.subplots(figsize=(10, 6))

sizes = [87, 13]

labels = ['Con cáncer', 'Sin cáncer']

ax.pie(sizes, labels=labels, autopct='%1.1f%%',
       colors = ['dodgerblue', 'lightblue'], textprops = {'fontsize':14})

plt.title("Distribución de los pacientes por condición de cáncer", fontsize = 20)
```

Con el gráfico resulta mucho más visible el porcentaje de pacientes con cáncer que forman parte de nuestro set de datos.

Ahora podría interesarnos observar cómo se distribuyen los mismos en cuanto a género y condición de cáncer, para adentrarnos un poco más a entender si existen patrones por género también. Primero observemos en una tabla los diferentes géneros que tenemos, de la siguiente manera:

```python
### Agrupo por género para observar los géneros y la cantidad de pacientes

generos = cancer_pulmon.groupby(by = 'genero').agg({'id':'count'}).reset_index()
generos.rename(columns = {'id':'cant_pacientes'}, inplace = True) ### Renombro la columna
generos['porcentaje'] = np.round((generos['cant_pacientes']/2000)*100, 1) ### Porcentaje del total
generos.sort_values(by = 'cant_pacientes', ascending = False) ### Ordeno de mayor a menor
```

Podemos observar que tenemos en total 8 géneros en nuestro set de datos, siendo casi el 90% hombres y mujeres. Lo que podemos hacer a continuación es agrupar a los demás géneros para facilitar los análisis, como se muestra a continuación:

```python
### Reemplazamos los géneros que representan poco volumen

cancer_pulmon.replace(['Non-binary', 'Polygender', 'Agender', 'Genderfluid',
                        'Bigender', 'Genderqueer'], 'Other', inplace = True)

generos1 = cancer_pulmon.groupby(by = 'genero').agg({'id':'count'}).reset_index()
generos1.rename(columns = {'id':'cant_pacientes'}, inplace = True) ### Renombro la columna
generos1['porcentaje'] = np.round((generos1['cant_pacientes']/2000)*100, 1) ### Porcentaje del total
generos1.sort_values(by = 'cant_pacientes', ascending = False) ### Ordeno de mayor a menor
```

Ahora podemos observar con mayor claridad los diferentes géneros, quedando así 3 grandes grupos: hombres, mujeres y otros, siendo este último una muestra mucho más importante para los análisis.

Lo siguiente que podemos observar es cómo se distribuyen estas cantidades entre los dos diagnósticos de los que disponemos.

```python
genero_diag = cancer_pulmon.groupby(by = ['genero', 'cancer_pulmon']).agg({'id':'count'}).reset_index()
genero_diag.rename(columns = {'id':'cant_pacientes', 'cancer_pulmon':'Cáncer de pulmón'}, inplace = True)
genero_diag.sort_values(by = 'cant_pacientes', ascending = False)

### Armamos un pivot con los datos para que resulte mejor visualmente
genero_diag_pivot = genero_diag.pivot(index = 'genero', columns = 'Cáncer de pulmón', values = 'cant_pacientes')
genero_diag_pivot.sort_values(['si', 'no'], ascending = False, inplace = True)
genero_diag_pivot.reset_index(inplace = True)
genero_diag_pivot
```

Con esta tabla podemos ver con mayor claridad la cantidad de pacientes por género y diagnóstico, siendo la mayoría con diagnóstico positivo como ya habíamos confirmado. Lo que nos muestran los datos es que por poco los que más propensos parecieran ser al cáncer son las mujeres, más que los hombres, al contrario de lo que planteamos en nuestras hipótesis.

Veamos estos datos graficados, para que sea más visual la interpretación de los mismos.

```python
### Reordenamos el dataframe para poder graficarlo
genero_diag.sort_values(by = 'cant_pacientes', ascending = False, inplace = True)
genero_diag.reset_index(drop = True, inplace = True)
### También podemos reemplazar los nombres de los géneros en inglés por los que son en español
genero_diag.replace(['Female', 'Male','Other'], ['Femenino', 'Masculino', 'Otro'], inplace = True)
```

```python
plt.figure(figsize = (10,6))

# paleta = {'lightblue', 'dodgerblue'}

bar1 = sns.barplot(x = 'genero', y = 'cant_pacientes', hue = 'Cáncer de pulmón',
                   data = genero_diag, palette = {'si':'dodgerblue', 'no':'lightblue'}, edgecolor = 'dimgray')

### Acá aplico los valores del pivot para agregar el valor a cada barra
plt.bar_label(bar1.containers[0], labels = genero_diag_pivot['si'], label_type = 'edge',
             fontsize=12)
plt.bar_label(bar1.containers[1], labels = genero_diag_pivot['no'], label_type = 'edge',
             fontsize=12)

plt.title("Cantidad de pacientes por género y diagnóstico", fontsize=16)
plt.xlabel(None)
plt.ylabel("Cantidad de pacientes", fontsize=10)
plt.tick_params(axis='both', labelsize=10)
plt.ylim(0, 900)
```

El gráfico nos permite observar mejor los datos y la información que nos proporcionan. Confirmamos con esto que los pacientes más propensos son mujeres pero los hombres también forman un volumen importante en los análisis, representando el 45% y el 44% respectivamente.

Lo siguiente que nos interesa saber es cómo se distribuyen los pacientes por edad, género y diagnóstico. Lo que podemos hacer para conocer esto es una distribución de densidad. A continuación se muestra cómo se logra esto:

```python
### En esta parte procedemos a separar los diferentes datos de los pacientes
### por género y diagnóstico

### Female
female = cancer_pulmon[cancer_pulmon['genero'] == 'Female']
female.reset_index(drop = True, inplace = True)
# Sin cáncer
female_no = female[female['cancer_pulmon'] == 'no']
female_no.reset_index(drop = True, inplace = True)
# Con cáncer
female_si = female[female['cancer_pulmon'] == 'si']
female_si.reset_index(drop = True, inplace = True)

### Male
male = cancer_pulmon[cancer_pulmon['genero'] == 'Male']
male.reset_index(drop = True, inplace = True)
# Sin cáncer
male_no = male[male['cancer_pulmon'] == 'no']
male_no.reset_index(drop = True, inplace = True)
# Con cáncer
male_si = male[male['cancer_pulmon'] == 'si']
male_si.reset_index(drop = True, inplace = True)

### Other
other = cancer_pulmon[cancer_pulmon['genero'] == 'Other']
other.reset_index(drop = True, inplace = True)
# Sin cáncer
other_no = other[other['cancer_pulmon'] == 'no']
other_no.reset_index(drop = True, inplace = True)
# Con cáncer
other_si = other[other['cancer_pulmon'] == 'si']
other_si.reset_index(drop = True, inplace = True)
```

```python
### Grafico los datasets por género y diagnóstico

fig, ax = plt.subplots(ncols = 3, figsize = (15, 5),
                       sharex = True, sharey = True)

fig.suptitle("Densidad por género y diagnóstico", fontsize = 16)

paleta = ['lightblue', 'dodgerblue']

### Female ###
sns.kdeplot(ax = ax[0], data = female_no['edad'], fill = True,
            color = paleta[0])
sns.kdeplot(ax = ax[0], data = female_si['edad'], fill = True,
            color = paleta[1])

ax[0].set_title("Femenino", fontsize = 12,
             loc = 'center')
ax[0].set_ylabel("Densidad")
ax[0].set_xlabel(None)

### Male ###
sns.kdeplot(ax = ax[1], data = male_no['edad'], fill = True,
            color = paleta[0])
sns.kdeplot(ax = ax[1], data = male_si['edad'], fill = True,
            color = paleta[1])

ax[1].set_title("Masculino", fontsize = 12,
             loc = 'center')
ax[1].set_ylabel(None)
ax[1].set_xlabel("Edad")

### Other ###
sns.kdeplot(ax = ax[2], data = other_no['edad'], fill = True,
            color = paleta[0], label = 'Sin cáncer')
sns.kdeplot(ax = ax[2], data = other_si['edad'], fill = True,
            color = paleta[1], label = 'Con cáncer')
ax[2].set_title("Otro", fontsize = 12, loc = 'center')
ax[2].set_xlabel(None)
ax[2].legend()
```

Viendo a los pacientes de esta manera se tiene un mejor entendimiento de las distribuciones, ya que si no destacabamos por género y condición de cáncer lo que ocurriría es que la distribución no destacaría los picos pronunciados que vemos en los gráficos.

Sabiendo esto y observando los gráficos de densidad se puede determinar que no hay diferencias significativas de los pacientes mujeres y hombres que presentan o no cáncer, ya que los gráficos casi se solapan, por lo que no hay distinción en cuanto a la edad en que un paciente de estos géneros puede contraer cáncer. Mientras que en el gráfico de 'other' se puede observar que existe un pico para los pacientes de ese grupo que no tienen cáncer y los que sí tienen cáncer se distribuyen un poco más de manera uniforme, presentando una leve pendiente a medida que crece la edad.

Lo que nos dice esto es que para los géneros femenino y masculino no hay un rango de edad tan definido como se planteó en las hipótesis sino que resulta ser bastante homogénea la edad en que los pacientes presentaron cáncer. Para los pacientes contenidos en la clase 'other' ocurre algo similar, los pacientes están contenidos dentro de un rango de edades que no destaca, a diferencia de los pacientes de esta clase que no tienen cáncer, los cuales generan un pico entre los rangos de edad 30 y 60 años.

Con esto podemos afirmar que la hipótesis es errónea, ya que los pacientes no respetan el rango de edad que se supuso en un inicio. También estos análisis responden a nuestra pregunta en cuanto a si existe o no una edad predilecta para contraer cáncer. Quizás calculando un promedio se puede estimar la edad pero no nos garantiza que sea de esa manera, ya que por lo visto los rangos de edades son bastante amplios.

Tenemos un archivo que nos ayudará a armar el mapa y a ubicar a los pacientes más adelante, en el cual tenemos los polígonos de cada país. Los polígonos son los que nos ayudan a generar una localicación en un mapa y son figuras geométricas hechas por puntos conectados entre sí, que delimitan los países en un mapa. Vamos a cargar el archivo para utilizarlo más tarde.

```python
### Nombro una variable que contenga la url donde se encuentra el archivo que necesito
file_countries = "C:/Users/175544/Desktop/Curso Coderhouse/Data Science/Avances PF/countries.geo.json"
```

Ahora, para poder graficar en un mapa lo que necesitamos es definir los polígonos del mapa pero como un DataFrame de geoPandas, por lo que lo transformamos a continuación.

```python
### Cargo el archivo como un geodataframe
geo_countries = gpd.read_file(file_countries)
geo_countries.rename(columns={'name':'pais','id':'abrev'},inplace=True)
geo_countries.head()
```

Ahora que ya tenemos armado los polígonos para el mapa lo que podemos hacer ahora es cruzar los países de los polígonos con los países de los pacientes.

```python
### Cruzo la base que tiene los polígonos con la base de pacientes a través de la columna 'pais'
cruce_dataset = pd.merge(geo_countries, cancer_pulmon, how='left', on='pais')
cruce_dataset.head()
```

Algo para pensar es que sabemos que entre padecimientos y diagnóstico podemos tener una relación, pero no entre las columnas como nombre, apellido, edad, etc, ya que estas representan una característica particular del paciente, por lo que estas columnas las eliminaremos para realizar este análisis.

A continuación se muestran los diferentes pasos para lograr esto.

```python
### Importamos esto para transformar las variables de texto en número
from sklearn import preprocessing
```

```python
### Armo el modelo de Encoding para las variables de texto
encoding = preprocessing.LabelEncoder()
```

```python
### Armo un nuevo set de datos en base al original
cancer_pulmon_heatmap = cancer_pulmon
cancer_pulmon_heatmap.drop(columns = ['id', 'nombre', 'apellido', 'genero',
                               'edad', 'pais'], inplace = True) ### Elimino las columnas que no necesito
```

```python
### Elimino las columnas que no presentan relación
cancer_heat = cancer_pulmon_heatmap

columnas = cancer_heat

for columna in columnas:

  # Aplico el modelo
  encoding.fit(cancer_heat[columna])
  list(encoding.classes_)

  # Transformo las variables
  cancer_heat[columna] = encoding.transform(cancer_heat[columna])

### Relleno los valores faltantes con ceros
cancer_heat.fillna(0)

### Armo la matriz de correlacion
matriz = cancer_heat.corr()

### Grafico la matriz en un heatmap
f, ax = plt.subplots(figsize=(15, 15))
heatmap = sns.heatmap(matriz,
                      # mask = mask,
                      square = True,
                      linewidths = .5,
                      cbar_kws = {'shrink': .4,
                                'ticks' : [-1, -.5, 0, 0.5, 1]},
                      vmin = -1,
                      vmax = 1,
                      annot = True,
                      annot_kws = {'size': 12},
                      cmap = 'Blues'
                    )

sns.set_style({'xtick.bottom': True}, {'ytick.left': True})
# plt.savefig('heatmap.jpg', dpi = 300)
```

Mediante el entrenamiento de estos modelos y utilizando métricas que nos permitan determinar qué modelo se ajusta mejor a nuestros datos podremos luego probar los mismos para otros set de datos con pacientes.

A continuación se desglosa una lista de pasos a seguir que resultan cruciales para poder conseguir un modelo óptimo.

Pero para realizar todo este proceso se deben tener en cuenta algunos pasos importantes.

1. Establecer la variable objetivo: será la variable que los modelos tendrán que predecir.
2. Tipo de modelo: se debe elegir entre regresión o clasificación, dependiendo de los datos.

#### Establecer variable objetivo

La variable que me interesa poder predecir por medio de modelos de Machine Learning es la de diagnóstico, donde nos dicen si el paciente en cuestión tiene cáncer o no, por lo que esta será la variable objetivo.

#### Tipo de modelo

Para trabajar mis datos utilizaré modelos de Clasificación ya que mi variable objetivo es categórica.

#### Relación entre los datos

Cuando se realizó el mapa de calor se pudo observar las relaciones que existen entre las diferentes variables del set de datos.

#### Selección de variables

Este paso también se realizó al armar el mapa de calor, siendo las elegidas las que se listan a continuación:



```python
### Genero un nuevo set de datos con lo utilizado en el mapa de calor
cancer_enc = cancer_heat

### Elijo las columnas que serán las variables
y_col = ['cancer_pulmon'] ### Variable objetivo

x_cols = ['fumador', 'dedos_amarillos', 'ansiedad', 'presion_social',
           'enfermedad_cronica', 'fatiga', 'alergias', 'sibilancia',
           'consume_alcohol', 'tos', 'dificultad_respiratoria',
           'dificultad_al_tragar', 'dolor_de_pecho'] ### Variables elegidas para los modelos
```

#### Prueba de modelos y métricas

Probamos a continuación los diferentes modelos de Clasificación de Machine Learning, que son: Árbol de decisión, KNN y Regresión Logística. Para esto tomamos las variables que se seleccionaron durante el armado de la matriz de correlación y entreno a los modelos. Luego para determinar la validez de los modelos se probarán las métricas F1 Score y la Matriz de Confusión para observar si las predicciones hechas por los modelos son correctas o no.

Entonces, para comenzar a realizar esto se importarán las librerías necesarias para esto.

```python
### Importo librerías necesarias para entrenar y para probar los modelos
# Para entrenar y testear
from sklearn.model_selection import train_test_split
# Árbol de decisión
from sklearn.tree import DecisionTreeClassifier
# Regresión Logística
from sklearn.linear_model import LogisticRegression
# Vecinos cercanos o KNN
from sklearn.neighbors import KNeighborsClassifier

#### Métricas
from sklearn.metrics import f1_score, confusion_matrix
from sklearn import metrics
from sklearn.model_selection import cross_val_score

### Importo la librería warnings para no tener carteles molestos
import warnings as wn
```

Armo dos pares de variables:

1. De entrenamiento (train): estas servirán para entrenar los modelos de ML.
2. De testeo (test): estas servirán para probar si al calcular las métricas obtengo overfitting o underfitting.

```python
### Genero los datos de entrenamiento para los modelos
Y = cancer_enc[y_col]

X = cancer_enc[x_cols]

### Variables de entrenamiento

x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                    test_size = 0.2, ## Porcentaje de los datos que será para testear los modelos
                                                    random_state = 50)
```

```python
##### Definición de los modelos
# Arbol de decisión
arbol = DecisionTreeClassifier(max_depth = 3)

# Regresión Logística
regresion = LogisticRegression(penalty = 'l2',
                               solver = 'liblinear',
                               multi_class = 'ovr')

# KNN
knn = KNeighborsClassifier()
```

```python
### Armo una lista con los modelos

modelos = [arbol, regresion, knn]
```

```python
### Creo un for que pruebe en bucle cada modelo y las métricas

for modelo in modelos:

    wn.filterwarnings('ignore')

    print(f'El modelo que usaremos a continuacion es {modelo} ->')

    # Entreno los modelos
    modelo.fit(x_train, y_train)

    # Definimos el valor de prediccion
    y_train_pred = modelo.predict(x_train)
    y_pred = modelo.predict(x_test)

    # Calculo las métricas con F1 Score
    f1_train = np.mean(cross_val_score(modelo, x_train, y_train, cv=10, scoring='f1'))
    f1_test = np.mean(cross_val_score(modelo, x_test, y_test, cv=10, scoring='f1'))

    # Imprimimos los valores
    print(f'La métrica F1 Score del modelo de entrenamiento es de: {f1_train}')
    print(f'La métrica F1 Score del modelo de testeo es de: {f1_test}')

    # Uso además la matriz de confusión
    cm = confusion_matrix(y_test, y_pred)

    # Graficamos la matriz de confusión
    sns.heatmap(cm,
                annot=True,
                fmt='g',
                xticklabels=['No', 'Sí'],
                yticklabels=['No', 'Sí'])
    plt.ylabel('Predicción',fontsize=13)
    plt.xlabel('Real',fontsize=13)
    plt.title('Matriz de confusión',fontsize=17)
    plt.show()
```

#### Interpretación de los resultados

Se puede observar en los tres modelos de ML que las métricas que se calcularon con las variables de entrenamiento son cercanas a 1 para los 3 casos, lo cual quiere decir que los modelos predicen muy bien la variable objetivo. Pero una cosa que sucede a veces cuando las métricas son cercanas a 1 es que hay overfitting, por lo que por esto se generaron variables de testeo y se predijeron de la misma manera, mostrando las siguientes comparaciones:

*Árbol de decisión:*
1. La métrica F1 Score del modelo de entrenamiento es de: 0.9359153576618736
2. La métrica F1 Score del modelo de testeo es de: 0.9376958334804393

*Regresión Logística:*
1. La métrica F1 Score del modelo de entrenamiento es de: 0.9653394659825529
2. La métrica F1 Score del modelo de testeo es de: 0.9575491985757326

*KNN:*
1. La métrica F1 Score del modelo de entrenamiento es de: 0.9771311270362408
2. La métrica F1 Score del modelo de testeo es de: 0.9604112695388153

A través de las métricas se observa que cualquiera de estos modelos predice adecuadamente los datos, ya que el modelo de entrenamiento y de testeo dan valores muy cercanos y además cercanos a 1. Pero si nos remontamos a la primera parte de este proyecto y recordamos que la mayoría de los pacientes de esta muestra efectivamente tenía cáncer, lo que podemos deducir con estos resultados es que los modelos predijeron bien pero porque la mayoría de las opciones que tenían para predecir eran positivas, por lo que puede afirmarse que los modelos "memorizaron" los valores de la muestra y por ende acertaron tanto para la muestra de entrenamiento como para la de testeo, y por esto dan las métricas unos resultados favorables.
